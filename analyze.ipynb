Aşağıdaki veriler:
- Obezite hastası erkekler
- Obezite hastası erkeklerde aynı zamanda diyabet olanlar

```python
male_obesity = data.loc[(data['Gender'] == 'Male') & (data['Obesity'] == 'Yes')]
male_positive = male_obesity.loc[male_obesity['class'] == 'Positive']
male_amount = male_positive.groupby('Age', as_index= False).size()
```

```python
male_obesity
```

```python
male_positive
```

```python
male_amount
```

Her yaş grubundan kaç tane hasta olduğu yukarıda yazıyor, bu verinin dictionary şeklinde sunumu:

```python
male_age_size = dict()
for age, size in male_amount.values:
    male_age_size[age] = size
```

```python
male_age_size
```

Aynı işlem aşağıda kadınlara da yapıldı.

```python
female_obesity = data.loc[(data['Gender'] == 'Female') & (data['Obesity'] == 'Yes')]
female_positive = female_obesity.loc[female_obesity['class'] == 'Positive']
female_amount = female_positive.groupby('Age', as_index= False).size()
```

```python
female_age_size = dict()
for age, size in female_amount.values:
    female_age_size[age] = size
```

```python
female_age_size
```

İlk Grafik:

```python
plt.subplot(2, 1, 1)
plt.suptitle('Obezite Hastalarında Yaşa Göre Diyabet Hastası Sayısı')

x_axis = [i for i in male_age_size.keys()]
y_axis = [i for i in male_age_size.values()]

plt.scatter(x_axis, y_axis, color= 'blue')
plt.legend(['Erkek'], loc= 'upper left')
plt.ylabel('Hasta Sayısı')
plt.subplot(2, 1, 2)

x_axis = [i for i in female_age_size.keys()]
y_axis = [i for i in female_age_size.values()]

plt.scatter(x_axis, y_axis, color= 'red')
plt.legend(['Kadın'], loc= 'upper left')
plt.xlabel('Hasta Yaşları')
plt.ylabel('Hasta Sayısı')
plt.show()
```

Hem halsizlik hem de kaşıntı sahibi hastalarla ilgili çalışma:

```python
male_weakness_itching = data.loc[(data['weakness'] == 'Yes') & (data['Itching'] == 'Yes') & (data['Gender'] == 'Male')]
all_male = male_weakness_itching.groupby(['Age'], as_index= False).size()['size'].sum()
```

```python
male_weakness_itching
```

```python
all_male
```

```python
male_positive = male_weakness_itching.loc[male_weakness_itching['class'] == 'Positive']
male_negative = male_weakness_itching.loc[male_weakness_itching['class'] == 'Negative']
```

```python
male_positive
```

```python
male_negative
```

```python
m_positive_amount = male_positive.groupby(['Age'], as_index= False).size()['size'].sum()
m_negative_amount = male_negative.groupby(['Age'], as_index= False).size()['size'].sum()
```

```python
print(f"Halsizlik ve Kaşıntı Hisseden Erkekler: {all_male}")
print(f"Diyabet Hastası Erkekler: {m_positive_amount}")
print(f"Diyabet Hastası Olmayan Erkekler: {m_negative_amount}")
print(f"Sağlama: {m_positive_amount} + {m_negative_amount} = {m_positive_amount + m_negative_amount}")
```

Kadınlar

```python
female_weakness_itching = data.loc[(data['weakness'] == 'Yes') & (data['Itching'] == 'Yes') & (data['Gender'] == 'Female')]
all_female = female_weakness_itching.groupby(['Age'], as_index= False).size()['size'].sum()
```

```python
female_weakness_itching
```

```python
all_female
```

```python
female_positive = female_weakness_itching.loc[female_weakness_itching['class'] == 'Positive']
female_negative = female_weakness_itching.loc[female_weakness_itching['class'] == 'Negative']
```

```python
female_positive
```

```python
female_negative
```

```python
f_positive_amount = female_positive.groupby(['Age'], as_index= False).size()['size'].sum()
f_negative_amount = female_negative.groupby(['Age'], as_index= False).size()['size'].sum()
```

```python
print(f"Halsizlik ve Kaşıntı Hisseden Kadınlar: {all_female}")
print(f"Diyabet Hastası Kadınlar: {f_positive_amount}")
print(f"Diyabet Hastası Olmayan Kadınlar: {f_negative_amount}")
print(f"Sağlama: {f_positive_amount} + {f_negative_amount} = {f_positive_amount + f_negative_amount}")
```

İkinci grafik:

```python
X = [m_positive_amount, m_negative_amount, f_positive_amount, f_negative_amount]
```

```python
plt.pie(X, colors= ['orange', 'lightgreen', 'darkmagenta', 'lightseagreen'], 
        labels= ['Erkek (+)', 'Erkek (-)', 'Kadın (+)', 'Kadın (-)'],
        autopct= "%1.2f%%",
        startangle= 90,
        wedgeprops= {'edgecolor': 'black',
                    'linewidth': 2,
                    'antialiased': True})
plt.title('Halsizlik ve Kaşıntı Hisseden Kadın-Erkek Hastaların Oransal Karşılaştırması')
plt.xlabel('(+): Diyabet Pozitif\n\n(—): Diyabet Negatif')

plt.show()
```
